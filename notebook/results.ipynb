```python
import json
from os.path import dirname, abspath
import os
import pandas as pd
```

```python
def open_json(filepath):
    with open(filepath, 'r') as f:
        data = json.load(f)
    return data
```

```python
f1_scores_path = dirname(abspath(os.getcwd())) + "/src/models/score/f1_scores.json"
f1_scores = open_json(f1_scores_path)

true_pos_wining_rates_path = dirname(abspath(os.getcwd())) + "/src/models/score/true_pos_wining_rates.json"
true_pos_wining_rates = open_json(true_pos_wining_rates_path)
```

```python
tuning_time = {'logis_reg': 901,
              'randomForest': 20352,
              'catboost': 5353,
              'lightGBM': 16369,
              'EN_HARD': 0,
              'EN_SOFT': 0}

training_time = {'logis_reg': 4.6,
              'randomForest': 26.3,
              'catboost': 4.0,
              'lightGBM': 4.5,
              'EN_HARD': 9.0,
              'EN_SOFT': 8.9}
```

```python
d = {
    "true_pos_wining_rates-training": true_pos_wining_rates["train"],
    "F1-score-training": f1_scores["train"],
    "true_pos_wining_rates-testing": true_pos_wining_rates["test"],
    "F1-score-testing": f1_scores["test"],
    "tuning_time": tuning_time,
    "training_time": training_time
}
algorithms_list = ["Logistic Regression", "Random Forest", "CatBoost", "LightGBM", "Hard Voting", "Soft Voting"]
df = pd.DataFrame(d).reset_index().drop(columns="index")
df.insert(0, "Algorithm", algorithms_list)
df.columns = ["Algorithm", "True Positive Winer - Train", "F1 score - Train",
             "True Positive Winer - Test", "F1 score - Test", "Tuning Time (seconds)", " Training Time (seconds)"]
df
```
